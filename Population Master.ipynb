```python
#Conditional Statement, custom function code from: https://blog.hubspot.com/website/pandas-if-else
def assign_popsize(row):

    if row <= 20:

        result = 'Largest nations'

    elif row <= 60:

        result = 'Large Nations'

    elif row <= 100:

        result = 'Mid Size Nations'

    elif row <= 200:

        result = 'Smaller Population'

    else:

        result = 'Smallest Population'

    return result

df2.loc['Population Description'] = df2['GlobalRank'].apply(assign_popsize)
```

```python
#Creating formula to find no. of non urban population, will use population with 10 mil plus citizens
a = df2['Population']
b = df2['UrbanPopulation']
non_urban = (a - b)
```

```python
#filtering for countries with over 10mil population in 2020 broken down by Urban and Non Urban
df3= df2[df2['Population'] >= 10000000]
new = df3[['Country', 'UrbanPopulation']].copy()
non_urban = pd.Series (non_urban)
new2 = new.assign(NonUrban = non_urban)
display(new2)
```
